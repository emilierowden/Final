```python
import pandas as pd
from matplotlib.pyplot import pie, axis, show
%matplotlib inline
df = pd.read_csv("/content/drive/MyDrive/Heart_disease_Final - Heart_disease_Final.csv")
sums = df.groupby(df["sex"])["num"].sum()
axis('equal');
labels= ['Female', 'Male']
#sums,labels=sums,
plt.pie(sums,labels=labels,  startangle=90, autopct='%1.1f%%')
plt.title('Percentage of Females and Males used in study')

show()
```
